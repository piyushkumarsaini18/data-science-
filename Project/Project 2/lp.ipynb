```python
from sklearn.preprocessing import LabelEncoder

# Step 1: Load the data
df = pd.read_csv("laptopPrice.csv")

```

```python
le = LabelEncoder()
cat_cols = [
    'brand', 'processor_brand', 'processor_name', 'processor_gnrtn',
    'ram_type', 'os', 'weight', 'warranty', 'Touchscreen', 'msoffice', 'rating'
]

#  Apply LabelEncoder
for col in cat_cols:
    df[col] = le.fit_transform(df[col].astype(str))

# Final check
print(df.head())
```

```python
df
```

```python
def clean_numeric(df, column):
    return (
        df[column]
        .astype(str)
        .str.strip()  # spaces hatao
        .str.replace("GB", "", regex=False)
        .str.replace("bit", "", regex=False)
        .str.replace(" ", "", regex=False)
        .str.replace("NotAvailable", "0", regex=False)
        .replace(["nan", "NaN", "None"], "0")  # string nan
        .fillna("0")                           # real NaN
        .apply(lambda x: ''.join(filter(str.isdigit, x)))  # sirf digits rakho
        .replace("", "0")                      # blank to 0
        .astype(int)
    )
```

```python
le = LabelEncoder()
df['ram_gb'] = clean_numeric(df, 'ram_gb')
df['ssd'] = clean_numeric(df, 'ssd')
df['hdd'] = clean_numeric(df, 'hdd')
df['graphic_card_gb'] = clean_numeric(df, 'graphic_card_gb')
df['os_bit'] = clean_numeric(df, 'os_bit')
```
