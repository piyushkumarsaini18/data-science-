### pie

```python
x=np.array([1,2,3,4])
plt.pie(x)
plt.show()
```

```python
x=np.array([1,2,3,4])

labels =["apple","banana","cheries","dates"]
myexplode =[0,0.2,0,0]

plt.pie(x,labels=labels,explode=myexplode)
plt.show()
```

```python
x=np.array([1,2,3,4])

labels =["apple","banana","cheries","dates"]
myexplode =[0,0.2,0,0]

plt.pie(x,labels=labels,explode=myexplode,startangle=90,shadow=True,colors=c)
plt.show()
```

```python
x=np.array([1,2,3,4])

labels =["apple","banana","cheries","dates"]
myexplode =[0,0.2,0,0]
c=['red','green','blue','yellow']

plt.pie(x,labels=labels,explode=myexplode,startangle=90,shadow=True,colors=c,autopct='%1.1f%%')
plt.legend(title = "my pie chart")
plt.show()
```
